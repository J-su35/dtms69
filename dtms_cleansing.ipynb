```python
df2['timestamp'] = pd.to_datetime(df2['timestamp'])
```

```python
s = pd.to_datetime(df2['timestamp']).dropna().drop_duplicates().sort_values()

start = s.min().floor('5min')
end   = s.max().ceil('5min') - pd.Timedelta(minutes=5)  

full = pd.date_range(start, end, freq='5min')
```

```python
day_count = pd.Series(s).groupby(pd.Series(s).dt.date).size()
incomplete_days = day_count[day_count < 384]
print(incomplete_days)
```

```python
# ถ้า period ตรง 1 เดือนพอดี และคาดหวัง 384 จุด/วัน
expected = 384 * day_count.shape[0]         # คิดจากจำนวนวันจริงที่พบ
actual   = len(s)
print("expected:", expected, "actual:", actual, "missing:", expected - actual)
```

```python
# 1 เดือน = 31 วัน → 2976 จุด
expected_fixed = 96 * 31
print("fixed_expected:", expected_fixed, "missing:", expected_fixed - actual)
```
